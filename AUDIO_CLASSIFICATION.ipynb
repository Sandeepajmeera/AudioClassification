```python
import librosa
import matplotlib.pyplot as plt
import IPython.display as ipd

# Corrected file path for Unix-like systems
filename = '/content/drive/MyDrive/UrbanSound8K/audio/fold1/103258-5-0-9.wav'

# Load the audio file
data, sample_rate = librosa.load(filename)

# Plot the waveform using matplotlib
plt.figure(figsize=(14, 5))
plt.plot(data)
plt.title('Waveform of Audio Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.show()

# Play the audio
ipd.Audio(filename)

```
